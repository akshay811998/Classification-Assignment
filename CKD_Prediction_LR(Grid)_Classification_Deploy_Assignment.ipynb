```python
import pickle
```

```python
loaded_model=pickle.load(open("finalized_model_logisticregression.sav",'rb'))
age = float(input("Enter age: "))
bp = float(input("Enter blood pressure: "))
sg_input = input("Enter specific gravity (a/b/c/d/e): ").lower()
sg_b, sg_c, sg_d, sg_e = 0, 0, 0, 0
if sg_input == 'b':
    sg_b = 1
elif sg_input == 'c':
    sg_c = 1
elif sg_input == 'd':
    sg_d = 1
elif sg_input == 'e':
    sg_e = 1
al = float(input("Enter albumin: "))
su = float(input("Enter sugar: "))
rbc = int(input("Enter rbc (normal=1, abnormal=0): "))
pc = int(input("Enter pc (normal=1, abnormal=0): "))
pcc = int(input("Enter pcc (present=1, notpresent=0): "))
ba = int(input("Enter ba (present=1, notpresent=0): "))
bgr = float(input("Enter blood glucose random: "))
bu = float(input("Enter blood urea: "))
sc = float(input("Enter serum creatinine: "))
sod = float(input("Enter sodium: "))
pot = float(input("Enter potassium: "))
hrmo = float(input("Enter hemoglobin: "))
pcv = float(input("Enter packed cell volume: "))
wc = float(input("Enter white blood cell count: "))
rc = float(input("Enter red blood cell count: "))
htn = int(input("Enter hypertension (yes=1, no=0): "))
dm = int(input("Enter diabetes mellitus (yes=1, no=0): "))
cad = int(input("Enter coronary artery disease (yes=1, no=0): "))
appet = int(input("Enter appetite (good=1, poor=0): "))
pe = int(input("Enter pedal edema (yes=1, no=0): "))
ane = int(input("Enter anemia (yes=1, no=0): "))
result = loaded_model.predict([[age, bp, sg_b, sg_c, sg_d, sg_e, al, su, rbc, pc, pcc, ba,
                                bgr, bu, sc, sod, pot, hrmo, pcv, wc, rc,
                                htn, dm, cad, appet, pe, ane]])
print(result)
if result == 0:
    print("No Chronic Kidney Disease Predicted")
else:
    print ("Yes, You're affected with Chronic Kidney Disease")
```
